```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.drop(columns=["Unnamed: 0", "id", "original_title"], axis = 1, inplace=True)
df.head()
```

```python
from datetime import date
```

```python
df['release_date'] = pd.to_datetime(df.release_date)
df.head()
```

```python
df['year'] = pd.DatetimeIndex(df['release_date']).year
df['month'] = pd.DatetimeIndex(df['release_date']).month
df['day'] = pd.DatetimeIndex(df['release_date']).day
df.head()
```

```python
# df_grouped = df.groupby(['year', 'title'])
df_group = df[['year', 'vote_count', 'vote_average', 'title']]
# df_grouped = df.groupby()['title']
df_group.head()
```

```python
def get_movie(year):
    top_movie = df_group[df_group['year'] == year].sort_values(by='vote_count', ascending=False).iloc[2]
    # for i in df_group.columns:
    print(f"Most popular movie of the year {year} is: ", top_movie['title'])

get_movie(2000)
```

```python
year = int(input("In which year do you want to see the highest grossing movies? (numbers only): "))

def perYear(year):
    movies = df_group[df_group["year"] == year].sort_values(by = "vote_count", ascending=False)
    top_movies = movies.head(10)
    print(f"The top 10 movies in the year {year} are: ")
    for i in top_movies['title']:
        print(i)
        
    print(top_movies)

perYear(year)
```
